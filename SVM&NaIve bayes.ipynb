```python
'''
3. What is the mathematical intuition behind SVM?
Core Idea:

The fundamental idea behind SVMs is to find the optimal hyperplane that maximizes the margin between different classes of data points. This hyperplane
acts as a decision boundary, separating the data into distinct categories.

Mathematical Formulation:

Hyperplane Equation: A hyperplane can be represented by the equation:
w . x + b = 0
where:

. w is the weight vector, perpendicular to the hyperplane.
. x is the input data point.
. b is the bias term, representing the offset of the hyperplane from the origin.
2. Margin: The margin is the distance between the hyperplane and the nearest data points of each class (support vectors). It is calculated as:
Margin = 2 / ||w||
where ||w|| is the Euclidean norm of the weight vector.

3. Optimization Objective: The goal of SVM is to find the hyperplane that maximizes the margin while ensuring correct classification of the data points.
   This can be expressed as an optimization problem:

Maximize Margin = 2 / ||w||
   Subject to: yi(w . xi + b) >= 1 for all i

where:

yi is the class label of the ith data point (-1 or +1).

xi is the ith data point.

This constraint ensures that all data points are correctly classified and lie on the correct side of the hyperplane with a margin of at least 1.

4. Lagrangian Formulation: To solve this optimization problem, we use the Lagrangian multiplier method, which introduces Lagrange multipliers (αi) for 
   each constraint:
L(w, b, α) = 1/2 ||w||^2 - Σ αi[yi(w . xi + b) - 1]
The goal is to minimize this Lagrangian function with respect to w and b, and maximize it with respect to αi.

5. Dual Problem: By taking the partial derivatives of the Lagrangian with respect to w and b and setting them to zero, we can derive the dual problem:
Maximize Σ αi - 1/2 Σ Σ αiαjyiyj(xi . xj)
   Subject to: Σ αiyi = 0 and αi >= 0 for all i

This dual problem is easier to solve and involves only the dot products of the data points (xi . xj), which allows for the use of kernel functions to
handle non-linearly separable data.

Kernel Trick:

For non-linearly separable data, the kernel trick is used to map the data to a higher-dimensional space where it can be linearly separated. The kernel
function calculates the dot product of the data points in this higher-dimensional space without explicitly transforming the data.

Intuition:

The mathematical intuition behind SVM is to find the hyperplane that maximizes the margin between classes, ensuring the best generalization performance
and robustness to noise. The use of Lagrange multipliers and the dual problem allows for efficient optimization and the kernel trick enables handling
non-linearly separable data.
'''
```

```python
'''
4. What is the role of Lagrange Multipliers in SVM?

Lagrange multipliers play a crucial role in SVMs by helping to solve the optimization problem of finding the optimal hyperplane that maximizes the 
margin between classes. They allow us to incorporate the constraints of the problem into the objective function and efficiently find the solution.

Here's a breakdown:

1. Constrained Optimization:
. The core objective of SVM is to find the hyperplane that maximizes the margin while ensuring all data points are correctly classified. This involves
a constrained optimization problem, where we want to maximize the margin subject to the constraint that all data points are on the correct side of the
hyperplane.

2. Introducing Lagrange Multipliers:
. To solve this constrained optimization problem, we use the method of Lagrange multipliers. This involves introducing a Lagrange multiplier (αi) for 
  each constraint, which acts as a penalty for violating the constraint.

3. Lagrangian Function:
The Lagrangian function combines the original objective function (maximizing the margin) with the constraints, using the Lagrange multipliers as 
weights:
L(w, b, α) = 1/2 ||w||^2 - Σ αi[yi(w . xi + b) - 1]
where:

. w is the weight vector of the hyperplane.
. b is the bias term.
. αi are the Lagrange multipliers.
. yi is the class label of the ith data point.
. xi is the ith data point.

4. Minimizing and Maximizing:
. The goal is to minimize the Lagrangian function with respect to w and b (to find the optimal hyperplane) and maximize it with respect to αi (to 
  enforce the constraints).
5. Dual Problem:
. By taking the partial derivatives of the Lagrangian with respect to w and b and setting them to zero, we can derive the dual problem:

Maximize Σ αi - 1/2 Σ Σ αiαjyiyj(xi . xj)
   Subject to: Σ αiyi = 0 and αi >= 0 for all i

This dual problem is easier to solve and involves only the dot products of the data points (xi . xj), which allows for the use of kernel functions to
handle non-linearly separable data.

6.Support Vectors:
The Lagrange multipliers (αi) associated with the support vectors (data points closest to the hyperplane) are non-zero, while those associated with 
other data points are zero. This means that only the support vectors contribute to the final solution, making SVMs memory efficient.

In summary:

Lagrange multipliers are essential in SVMs because they allow us to:

. Incorporate constraints into the optimization problem.
. Derive the dual problem, which is easier to solve.
. Identify support vectors, which are crucial for defining the hyperplane.
'''
```

```python
'''
5. What are Support Vectors in SVM?

Support vectors are the data points that lie closest to the decision boundary (hyperplane) in an SVM model. They are the most critical elements in 
determining the optimal hyperplane and have a direct influence on its position and orientation.

Here's a breakdown:

1.Importance:
. Support vectors are the data points that are most difficult to classify, as they lie closest to the decision boundary between classes.
. They have the most influence on the position and orientation of the hyperplane, as even small changes in their position can significantly affect the
  decision boundary.
. They are the only data points that matter for defining the hyperplane; other data points can be removed without affecting the model's performance.

2. Identification:
. Support vectors are identified during the training process of an SVM model.
. They are the data points for which the corresponding Lagrange multipliers (αi) are non-zero.
. This means that only the support vectors contribute to the final solution, making SVMs memory efficient.

3. Role in Defining the Hyperplane:
. The hyperplane is defined by the support vectors and their corresponding Lagrange multipliers.
. The weight vector (w) of the hyperplane is a linear combination of the support vectors:

w = Σ αiyi xi
where:

. αi is the Lagrange multiplier for the ith support vector.
. yi is the class label of the ith support vector.
. xi is the ith support vector.
. The bias term (b) is also determined by the support vectors.

4. Margin:
. The margin is the distance between the hyperplane and the nearest support vectors.
. Maximizing the margin is the primary objective of SVM, as it leads to better generalization performance and robustness to noise.
. The support vectors define the margin and play a crucial role in its maximization.

In simpler terms:

Imagine you're trying to separate two groups of objects with a straight line. The support vectors would be the objects closest to the line, and they 
would determine the position of the line. If you move any of the other objects, the line wouldn't change, but if you move a support vector, the line
would have to move as well.

In summary:

Support vectors are the critical data points in an SVM model that:

. Lie closest to the decision boundary.
. Have the most influence on the hyperplane's position and orientation.
. Define the margin and contribute to its maximization.
. Are identified by non-zero Lagrange multipliers.
'''
```

```python
'''
8. What is the Kernel Trick in SVM?
The Kernel Trick

The Kernel Trick is a technique used in SVMs to handle non-linearly separable data. It allows SVMs to implicitly map data points to a higher-dimensional
space where they become linearly separable, without actually performing the computationally expensive transformation.

How it works:

1. Non-linear Data: When data points are not linearly separable in their original feature space, it means that a straight line cannot effectively 
   separate them into different classes.
2. Mapping to Higher Dimensions: The Kernel Trick avoids explicitly transforming the data to a higher-dimensional space by using a kernel function. The
   kernel function calculates the dot product of two data points in the higher-dimensional space without actually performing the transformation.
3. Implicit Transformation: This implicit transformation allows SVMs to find a hyperplane that separates the data in the higher-dimensional space, which
   corresponds to a non-linear decision boundary in the original feature space.
In simpler terms:

Imagine you have a dataset that cannot be separated by a straight line. The Kernel Trick acts like a magic trick that projects the data into a 
higher-dimensional space where it becomes separable by a plane. However, it never actually performs the projection; it just calculates the relationships
between data points as if they were in the higher-dimensional space.

Benefits of the Kernel Trick:
. Efficiency: Avoids the computational cost of explicitly transforming data to higher dimensions.
. Flexibility: Allows SVMs to handle non-linear relationships in data.
. Versatility: Can be used with various kernel functions to suit different data distributions.

Common Kernel Functions:

. Linear Kernel: Used for linearly separable data.
. Polynomial Kernel: Creates polynomial combinations of features.
. Radial Basis Function (RBF) Kernel: Measures similarity between data points based on distance.
Sigmoid Kernel: Introduces non-linearity using a sigmoid function.
In summary:

The Kernel Trick is a powerful technique that enables SVMs to handle non-linear data by implicitly mapping it to a higher-dimensional space. It is a 
crucial component of SVMs and contributes to their effectiveness in various machine learning tasks.
'''
```

```python
'''
12. What is the Naïve Bayes classifier, and why is it called "Naïve"?
What is the Naïve Bayes classifier?

The Naïve Bayes classifier is a probabilistic machine learning algorithm based on Bayes' Theorem. It is used for classification tasks, such as spam 
detection, sentiment analysis, and document categorization.

How it works:

1. Assumption: The Naïve Bayes classifier makes a crucial assumption called the "independence assumption." It assumes that the presence or absence of a 
   particular feature is unrelated to the presence or absence of any other feature, given the class variable.

2. Bayes' Theorem: It uses Bayes' Theorem to calculate the probability of a data point belonging to a particular class, given its features. The formula
   is:
P(class|features) = [P(features|class) * P(class)] / P(features)
Where:

. P(class|features): Posterior probability - the probability of a data point belonging to a specific class, given its features.
. P(features|class): Likelihood - the probability of observing the features, given the class.
. P(class): Prior probability - the probability of a data point belonging to a specific class.
. P(features): Evidence - the probability of observing the features.

3. Prediction: The classifier predicts the class with the highest posterior probability for a given data point.
Why is it called "Naïve"?

The classifier is called "naïve" because of its strong independence assumption. In real-world datasets, features are often correlated or dependent on 
each other. However, the Naïve Bayes classifier ignores these dependencies, simplifying the calculations and making the algorithm computationally 
efficient.

Despite its simplicity, Naïve Bayes often performs surprisingly well in practice, even when the independence assumption is violated. This is because
the classifier focuses on the relative probabilities of features within each class, which can still be informative for classification.

Example:

Consider classifying an email as spam or not spam based on the words it contains. The Naïve Bayes classifier would assume that the presence of words
like "free," "money," and "prize" are independent of each other, given the class (spam or not spam). Even though these words might often appear 
together in spam emails, the classifier treats them as separate pieces of evidence.

In code (using scikit-learn in Python):
from sklearn.naive_bayes import GaussianNB

# Create a Naïve Bayes classifier
classifier = GaussianNB()

# Train the classifier
classifier.fit(X_train, y_train)

# Make predictions
y_pred = classifier.predict(X_test)

'''
```

```python
'''
13. What is Bayes' Theorem?
Bayes' Theorem is a fundamental concept in probability theory and statistics that describes how to update the probability of a hypothesis based on new
evidence. It provides a way to revise existing beliefs in light of new information.

Formula:
P(H|E) = [P(E|H) * P(H)] / P(E)

Where:

. P(H|E): Posterior probability - the probability of hypothesis H being true, given evidence E.
. P(E|H): Likelihood - the probability of observing evidence E, given hypothesis H is true.
. P(H): Prior probability - the initial probability of hypothesis H being true, before considering any evidence.
. P(E): Evidence - the probability of observing evidence E.

In simpler terms:

Bayes' Theorem tells us how to adjust our beliefs about something (hypothesis) after seeing new data (evidence). It combines our prior knowledge with 
the likelihood of the evidence to arrive at a revised probability (posterior probability).

Example:

Let's say you want to know the probability of having a disease (H) given that you tested positive for it (E).

. P(H): Prior probability - the prevalence of the disease in the population (e.g., 0.01 or 1%).
. P(E|H): Likelihood - the probability of testing positive if you have the disease (e.g., 0.95 or 95%).
. P(E): Evidence - the probability of testing positive, which includes both true positives and false positives.

Using Bayes' Theorem, you can calculate P(H|E), the probability of having the disease given a positive test result. This updated probability reflects
both the initial prevalence of the disease and the accuracy of the test.

Applications:

Bayes' Theorem has numerous applications in various fields, including:

. Medical diagnosis: Assessing the probability of a disease based on symptoms and test results.
. Spam filtering: Classifying emails as spam or not spam based on the words they contain.
. Machine learning: Building probabilistic models for prediction and classification tasks, such as Naïve Bayes classifiers.
. Finance: Modeling risk and making investment decisions.

Key takeaways:

. Bayes' Theorem provides a framework for updating beliefs based on new evidence.
. It combines prior knowledge with the likelihood of the evidence to arrive at a revised probability.
. It has wide-ranging applications in various fields.
'''
```

```python
'''
16. What are the key assumptions made by Naïve Bayes?

The Naïve Bayes classifier, despite its effectiveness, relies on a few crucial assumptions. These assumptions, while often unrealistic in real-world
scenarios, contribute to its simplicity and computational efficiency. Here are the key assumptions:

1. Feature Independence:

. Assumption: The most important assumption is that features are conditionally independent of each other, given the class label. This means that the
  presence or absence of one feature does not affect the presence or absence of any other feature, once we know the class to which the data point
  belongs.

. Example: In spam detection, the presence of the word "free" is assumed to be independent of the presence of the word "money," given that the email is
  spam or not spam.

. Reality: This assumption is often violated in practice, as features in real-world datasets tend to be correlated or dependent on each other. However,
  Naïve Bayes often performs surprisingly well even with this violation.

2. Feature Importance (Equal Importance):

. Assumption: All features contribute equally to the outcome. This implies that the classifier treats each piece of information as having equal weight
  in determining class membership.
. Reality: In many cases, some features may be more informative or relevant than others. Ignoring feature importance may lead to suboptimal results if 
  certain features are much more relevant for a specific classification task.
. Example: In sentiment analysis, the presence of words like "fantastic" or "terrible" might carry more weight compared to neutral words like "the" or
  "a."
3. Data Distribution (depending on the variant):

. Gaussian Naïve Bayes: Assumes that continuous features follow a normal (Gaussian) distribution within each class.
. Multinomial Naïve Bayes: Assumes that features represent discrete counts or frequencies and follow a multinomial distribution.
. Bernoulli Naïve Bayes: Assumes that features are binary (present or absent) and follow a Bernoulli distribution.

Why Naïve Bayes works despite these assumptions:

. Focus on relative probabilities: Naïve Bayes primarily focuses on the relative probabilities of features within each class. Even if the independence 
assumption is violated, these relative probabilities can still be informative for classification.
. Error cancellation: In some cases, the errors introduced by the independence assumption can cancel each other out, leading to reasonable performance.
. Simplicity and speed: The assumptions make the algorithm computationally efficient and easy to implement, which can be advantageous in many 
  applications.
'''
```

```python
'''
17. What are the advantages and disadvantages of Naïve Bayes?
Advantages:

1. Simplicity and Speed: Naïve Bayes is easy to understand and implement. It's computationally efficient, requiring less training time and resources
   compared to more complex algorithms. This makes it suitable for large datasets and real-time applications.

2. Good Performance with High-Dimensional Data: It performs well even with high-dimensional data (many features), as it doesn't suffer from the curse
   of dimensionality as much as some other algorithms.

3. Handles Missing Data Well: Naïve Bayes can handle missing data by simply ignoring the missing values during probability calculations.

4.Works with Categorical and Numerical Data: While different variants are tailored to specific data types (Gaussian for numerical, Multinomial for
 categorical), Naïve Bayes can generally handle both types of features.

5. Good for Text Classification: Multinomial and Bernoulli Naïve Bayes are particularly effective for text classification tasks like spam filtering and
   sentiment analysis.

6. Requires Less Training Data: Compared to more complex models, Naïve Bayes can achieve good performance with relatively less training data.

7. Easy to Interpret: The results of Naïve Bayes are relatively easy to interpret, as you can examine the probabilities associated with different 
   features and classes.

Disadvantages:

1. Strong Independence Assumption: The "naïve" assumption of feature independence is often unrealistic in real-world scenarios. This can limit the 
   accuracy of the classifier when features are strongly correlated.

2. Zero Frequency Problem: If a feature value is not observed in the training data for a particular class, the classifier assigns a zero probability to
   it. This can lead to incorrect predictions. Techniques like Laplace smoothing are used to mitigate this issue.

3. Sensitivity to Irrelevant Features: Naïve Bayes can be sensitive to irrelevant features, which can negatively impact its performance. Feature 
   selection techniques can help address this problem.

4. Limited Expressiveness: Due to its simplicity, Naïve Bayes might not capture complex relationships between features and classes as effectively as
   more sophisticated algorithms.

5. Potential for Overfitting: While generally less prone to overfitting than some models, Naïve Bayes can still overfit if the training data is noisy
   or biased.

6. May struggle in case of data scarcity: Some variants, such as Gaussian Naïve Bayes, may be sensitive to small numbers of training points, or sparse
   data.
'''
```

```python
'''
19. Compare SVM and Naïve Bayes for classification tasks.

Both Support Vector Machines (SVM) and Naïve Bayes are popular algorithms for classification, but they differ in their underlying principles and 
strengths. Here's a comparison:

Support Vector Machines (SVM):

Principle: Aims to find the optimal hyperplane that maximally separates data points of different classes. It focuses on the points closest to the
decision boundary (support vectors) and tries to maximize the margin between them.
. Strengths:
. Effective in high-dimensional spaces.
. Relatively memory efficient.
. Versatile due to the use of different kernel functions (linear, polynomial, RBF).
. Robust to outliers due to the focus on support vectors.

. Weaknesses:
. Can be computationally expensive for large datasets.
. Sensitive to the choice of kernel and hyperparameters.
. Less interpretable than Naïve Bayes.

. Suitable for:
   . Complex datasets with non-linear relationships.
   . Cases where maximizing the margin between classes is important.
   . Tasks where interpretability is not a primary concern.

Naïve Bayes:

. Principle: Based on Bayes' Theorem, it calculates the probability of a data point belonging to a particular class, given its features. It assumes 
 feature independence, which simplifies the calculations.

. Strengths:
   . Simple and easy to implement.
   . Computationally efficient, especially for large datasets.
   . Handles high-dimensional data well.
   . Works with both categorical and numerical data.
   . Relatively insensitive to irrelevant features.
   . Good for text classification and other discrete data problems.
   Provides probabilities for predictions, allowing for uncertainty estimation.

. Weaknesses:
  . The independence assumption can be unrealistic.
  . Sensitive to the presence of features not seen in the training data (zero frequency problem).
  . May not capture complex relationships between features as effectively as SVM.

. Suitable for:
    . Text classification, spam filtering, sentiment analysis.
    . Problems with discrete or categorical features.
    . Cases where speed and simplicity are priorities.
    . Tasks where interpretability and uncertainty estimation are important.
'''
```

```python
'''
20. How does Laplace Smoothing help in Naïve Bayes?
Laplace Smoothing, also known as add-one smoothing, is a technique used in Naïve Bayes to address the zero frequency problem.

The Zero Frequency Problem:

In Naïve Bayes, we calculate probabilities based on the frequency of features in the training data. However, if a particular feature value is not 
observed in the training data for a specific class, the classifier assigns a zero probability to it. This can lead to issues:

. Incorrect Predictions: If a new data point contains this unseen feature value, the classifier might incorrectly predict its class, as the probability
  calculation becomes zero.
. Bias towards Frequent Features: The classifier might become overly biased towards features that are frequently observed in the training data, even if
  they are not truly representative of the underlying distribution.
How Laplace Smoothing Helps:

Laplace smoothing addresses the zero frequency problem by adding a small constant (usually 1) to the count of each feature value, including those that have not been observed in the training data. This ensures that no feature value has a zero probability.

Formula:

The smoothed probability for a feature value f in class c is calculated as:
P(f|c) = (count(f, c) + 1) / (count(c) + V)

Where:

. count(f, c): The number of times feature f appears in class c in the training data.
. count(c): The total number of data points in class c.
. V: The number of possible values for feature f (vocabulary size for text data).
'''
```

```python
'''
24. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision
boundary.
'''
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

# Load the Iris dataset (using only two features for visualization)
iris = datasets.load_iris()
X = iris.data[:, :2]  # Take only the first two features (sepal length and sepal width)
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create an SVM classifier with a polynomial kernel
classifier = SVC(kernel='poly', degree=3)  # degree is the order of the polynomial

# Train the classifier
classifier.fit(X_train, y_train)

# Create a meshgrid to plot the decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))

# Predict the class for each point in the meshgrid
Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary and data points
plt.contourf(xx, yy, Z, alpha=0.8)  # Filled contours for decision regions
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')  # Scatter plot of data points
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('SVM with Polynomial Kernel')
plt.show()
```

```python
'''
33. Write a Python program to implement a Naïve Bayes classifier for spam detection using email data.
'''
import re
import numpy as np
from collections import Counter
from sklearn.model_selection import train_test_split

def preprocess_email(email):
    """Preprocesses an email by converting to lowercase, removing punctuation, and splitting into words."""
    email = email.lower()
    email = re.sub(r'[^\w\s]', '', email)  # Remove punctuation
    words = email.split()
    return words

def train_naive_bayes(emails, labels):
    """Trains a Naïve Bayes classifier for spam detection."""
    num_emails = len(emails)
    num_spam = sum(labels)
    num_ham = num_emails - num_spam

    # Calculate prior probabilities
    prior_spam = num_spam / num_emails
    prior_ham = num_ham / num_emails

    # Create vocabulary and word counts for spam and ham emails
    spam_words = []
    ham_words = []
    for i, email in enumerate(emails):
        words = preprocess_email(email)
        if labels[i] == 1:  # Spam
            spam_words.extend(words)
        else:  # Ham
            ham_words.extend(words)

    spam_word_counts = Counter(spam_words)
    ham_word_counts = Counter(ham_words)
    vocabulary = set(spam_words + ham_words)

    # Calculate conditional probabilities (with Laplace smoothing)
    cond_probs_spam = {}
    cond_probs_ham = {}
    for word in vocabulary:
        cond_probs_spam[word] = (spam_word_counts[word] + 1) / (len(spam_words) + len(vocabulary))
        cond_probs_ham[word] = (ham_word_counts[word] + 1) / (len(ham_words) + len(vocabulary))

    return prior_spam, prior_ham, cond_probs_spam, cond_probs_ham, vocabulary

def predict_naive_bayes(email, prior_spam, prior_ham, cond_probs_spam, cond_probs_ham, vocabulary):
    """Predicts whether an email is spam or ham using the trained Naïve Bayes classifier."""
    words = preprocess_email(email)
    prob_spam = prior_spam
    prob_ham = prior_ham
    for word in words:
        if word in vocabulary:
            prob_spam *= cond_probs_spam[word]
            prob_ham *= cond_probs_ham[word]

    return 1 if prob_spam > prob_ham else 0  # 1 for spam, 0 for ham

# Sample email data and labels (replace with your actual data)
emails = [
    "Hey, check out this amazing deal!",
    "Important information about your account.",
    "Get free viagra now!",
    "Meeting reminder: tomorrow at 10 AM.",
    "Your package has been delivered.",
]
labels = [1, 0, 1, 0, 0]  # 1 for spam, 0 for ham

# Train the Naïve Bayes classifier
prior_spam, prior_ham, cond_probs_spam, cond_probs_ham, vocabulary = train_naive_bayes(emails, labels)

# Predict on a new email
new_email = "Congratulations! You have won a prize."
prediction = predict_naive_bayes(new_email, prior_spam, prior_ham, cond_probs_spam, cond_probs_ham, vocabulary)

if prediction == 1:
    print("The email is classified as spam.")
else:
    print("The email is classified as ham.")
```

```python
'''
46. Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.
'''
import matplotlib.pyplot as plt
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris  # Or any classification dataset
from sklearn.metrics import precision_recall_curve, average_precision_score
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier

# Load the Iris dataset (or any classification dataset)
iris = load_iris()
X = iris.data
y = iris.target

# Binarize the target labels for multi-class (One-vs-Rest)
y = label_binarize(y, classes=[0, 1, 2])
n_classes = y.shape[1]

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the SVM classifier using OneVsRestClassifier for multi-class
classifier = OneVsRestClassifier(SVC(kernel='linear', probability=True, random_state=42))  # probability=True for decision_function
classifier.fit(X_train, y_train)

# Get prediction probabilities or decision function for each class
y_score = classifier.decision_function(X_test)

# Compute Precision-Recall and plot curve
precision = dict()
recall = dict()
average_precision = dict()

for i in range(n_classes):
    precision[i], recall[i], _ = precision_recall_curve(y_test[:, i], y_score[:, i])
    average_precision[i] = average_precision_score(y_test[:, i], y_score[:, i])

# A "micro-average": quantifying score on all classes jointly
precision["micro"], recall["micro"], _ = precision_recall_curve(y_test.ravel(), y_score.ravel())
average_precision
```
